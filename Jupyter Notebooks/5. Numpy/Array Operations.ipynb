```python
import numpy as np

array1 = np.array([1,2,3,4,5])
array2 = np.array([6,7,8,9,10])

sum = array1+array2
print(sum)
```

```python
sub = array1 - array2

print(sub)
```

```python
mul = array1*array2
print(mul)
```

```python
div = array1/array2
print(div)
```

```python
mean = np.mean(array1)
print(mean)
```

```python
sum = np.sum(array1)
print(sum)
```

```python
max = np.max(array1)
min = np.min(array1)

print(min)
print(max)
```

```python
greater_than = array1 >3

print(greater_than)
```
